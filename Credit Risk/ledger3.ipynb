```python
dataset.columns =['Date','particulars','sales', 'vch type','vch_number','debit','credit'] 
dataset.head(50)
```

```python
dataset1 = dataset.iloc[10:]
dataset1.head(50)
```

```python
dataset1['customer_name']=dataset1['particulars']
dataset1.head(50)
```

```python
dataset1=dataset1.replace(to_replace ="To", value =np.nan) 
dataset1=dataset1.replace(to_replace ="By", value =np.nan) 
dataset1=dataset1.replace(to_replace ="Particulars", value =np.nan) 
dataset1.head(50)
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
dataset1.head(50)
```

```python
cols = ['customer_name']
dataset1.loc[:,cols] = dataset1.loc[:,cols].ffill()
dataset1.head(50)
```

```python
dataset1 = dataset1[dataset1.Date != 'Ledger:']
dataset1 = dataset1[dataset1.Date !='Date']
dataset1.head(50)
```

```python
dataset1=dataset1.drop(['particulars'], axis = 1) 
dataset1.head(50)
```

```python
dataset1['credit'] = dataset1['credit'].fillna(0)
dataset1['debit'] = dataset1['debit'].fillna(0)
```

```python
dataset1=dataset1.drop(['vch type'], axis = 1) 
dataset1=dataset1.drop(['vch_number'], axis = 1) 
```

```python
dataset1['Date'] = pd.to_datetime(dataset1['Date'], errors='coerce')
dataset1.head(50)
```

```python
dataset1=dataset1.dropna(subset=['sales'])
dataset1.head(50)
```

```python
cols = ['Date']
dataset1.loc[:,cols] = dataset1.loc[:,cols].ffill()
dataset1.head(50)
```

```python
dataset1['sales'] = dataset1['sales'].replace(['GST SALES', 'DISCOUNT ALLOWED', 'CASH','SOUTH INDIAN BANK LTD (CC A/C NO. 0038084000002275)', 'HDFC BANK'],['NP','PAYMENT','PAYMENT','PAYMENT','PAYMENT'])
```

```python
dataset1.reset_index(inplace = True, drop = True)
dataset1.head(50)
```

```python
dataset1['Gross Total']=dataset1['credit']-dataset1['debit']
```

```python
dataset1=dataset1.drop(['debit'], axis = 1) 
dataset1=dataset1.drop(['credit'], axis = 1)
```

```python
dataset1['days'] = dataset1.groupby('customer_name')['Date'].diff().apply(lambda x: x.days)
```

```python
dataset1['balance'] = dataset1.groupby('customer_name')['Gross Total'].cumsum()
dataset1.head(50)
```

```python
dataset1['days'] = dataset1['days'].fillna(0)
```
