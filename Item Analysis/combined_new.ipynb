```python
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df_cust_combined = pd.read_csv('calicut_item_comb.csv')
df_cust_combined
```

```python
df_merge_cust = df_cust_combined
```

```python
scaler = MinMaxScaler()
df_merge_cust[["Profit", "Frequency"]] = scaler.fit_transform(df_merge_cust[["profit_final", "frequency_final"]])
```

```python
df_merge_cust['Sales Score'] = 4*df_merge_cust['Profit'] + df_merge_cust['Frequency']
df_merge_cust[["Sales Score"]] = scaler.fit_transform(df_merge_cust[["Sales Score"]])
```

```python
df_merge_cust['Profit'] = df_merge_cust['Profit']*100
df_merge_cust['Frequency'] = df_merge_cust['Frequency']*100
df_merge_cust['Sales Score'] = df_merge_cust['Sales Score']*100
```

```python
df_merge_cust.describe()
```

```python
df_test = df_merge_cust[df_merge_cust['Frequency'] > 80 ]
df_test
```

```python
df_merge_cust['Sales Segment'] = 'Low-Value'
df_merge_cust.loc[df_merge_cust['Sales Score'] >= 15,'Sales Segment'] = 'Below-Average' 
df_merge_cust.loc[df_merge_cust['Sales Score'] >= 50,'Sales Segment'] = 'Average' 
df_merge_cust.loc[df_merge_cust['Sales Score'] >= 80,'Sales Segment'] = 'High-Value' 
```

```python
df_merge_cust['Profit Segment'] = 'Low-Value'
df_merge_cust.loc[df_merge_cust['Profit'] >= 15,'Profit Segment'] = 'Below-Average' 
df_merge_cust.loc[df_merge_cust['Profit'] >= 50,'Profit Segment'] = 'Average' 
df_merge_cust.loc[df_merge_cust['Profit'] >= 80,'Profit Segment'] = 'High-Value' 
```

```python
df_merge_cust['Frequency Segment'] = 'Low-Value'
df_merge_cust.loc[df_merge_cust['Frequency'] >= 15,'Frequency Segment'] = 'Below-Average' 
df_merge_cust.loc[df_merge_cust['Frequency'] >= 50,'Frequency Segment'] = 'Average' 
df_merge_cust.loc[df_merge_cust['Frequency'] >= 80,'Frequency Segment'] = 'High-Value' 
```

```python
df_merge_cust.drop(["profit_final", "frequency_final", "sales_score"], axis = 1, inplace = True)
```

```python
df_merge_cust.rename(columns = {'customer_name':'Customer Name'}, inplace = True)
```

```python
df_cust_quarter = pd.read_csv('calicut_item_quarterwise.csv')
df_cust_quarter
```

```python
df_cust_quarter.sort_values("QY", axis = 0, ascending = True, 
                 inplace = True, na_position ='last') 
qy_val = df_cust_quarter.QY.unique() 
qy_val
```

```python
l = len(qy_val)
i=0
while(i<l):
    globals()['df_cust_quarter%s' % i] = df_cust_quarter[df_cust_quarter['QY']==qy_val[i]]
    i = i+1
```

```python
i = 0
while (i<l):
    
    globals()['df_cust_quarter%s' % i][['frequency_cluster','profit_cluster']] = scaler.fit_transform(globals()['df_cust_quarter%s' % i][['frequency_cluster','profit_cluster']])
    
    i= i+1
```

```python
i = 0
list_of_dataframes = []
while(i<l):
    list_of_dataframes.append(globals()['df_cust_quarter%s' % i])
    i = i+1
```

```python
df_cust_quarter_final = pd.concat(list_of_dataframes)
```

```python
df_cust_quarter_final['Frequency Score'] = df_cust_quarter_final['frequency_cluster']*100
df_cust_quarter_final['Profit Score'] = df_cust_quarter_final['profit_cluster']*100
```
